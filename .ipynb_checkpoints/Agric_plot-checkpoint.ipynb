```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
map = Basemap(llcrnrlon=33.575667,llcrnrlat=-5.592822,urcrnrlon=42.079084,urcrnrlat=5.42044,
             resolution="h",projection='mill', lat_0=0.133138,lon_0=37.794417)
map.drawmapboundary(fill_color="aqua")
map.fillcontinents(color="#EDEBC8", lake_color="aqua")
map.drawcountries()
map.drawcoastlines()
map.readshapefile('/home/paul/Ke_Agric_density/ke_agriculture/ke_agriculture','ke_agriculture',color='green')
map.readshapefile('/home/paul/Ke_Agric_density/kenyan-counties/County','County',color="#F16C10")

plt.title('KENYA AGRICULTURAL DENSITY MAP',fontweight='bold',fontsize='15',color='orange')
plt.savefig('ke_agric.png', bbox_inches='tight')
```
